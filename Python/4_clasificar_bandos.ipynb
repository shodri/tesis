```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('comentarios.csv')
```

```python

# Calcular el número de ocurrencias de cada valor en la columna 'Bando'
counts = df['Bando'].value_counts()

# Mapear las etiquetas de las barras a los valores deseados y definir el orden deseado
etiquetas = {
    'A': 'A favor',
    'C': 'En contra',
    'N': 'Neutrales'
}

# Definir el orden deseado de las etiquetas
orden_etiquetas = ['A favor', 'En contra', 'Neutrales']

# Reemplazar las etiquetas en el gráfico y reordenar según el orden deseado
counts.index = counts.index.map(etiquetas)
counts = counts.reindex(orden_etiquetas)

# Graficar la diferencia entre los valores 'A' y 'C'
plt.bar(counts.index, counts.values, color=['green', 'red', 'grey'])
plt.xlabel('Posición')
plt.ylabel('Número de comentarios')
plt.title('Comentarios a favor, en contra o neutrales al proyecto')
plt.show()
```

```python

# Obtener los porcentajes de cada bando
porcentajes = df['Bando'].value_counts(normalize=True) * 100

# Definir los colores para cada bando
colors = {'A': 'green', 'C': 'red', 'N': 'grey'}
etiquetas = {
    'C': 'En contra',
    'A': 'A favor',
    'N': 'Neutrales'
}
# Crear el gráfico de torta con colores personalizados

plt.pie(porcentajes, labels=etiquetas.values(), autopct='%1.1f%%', colors=[colors[key] for key in porcentajes.index])
plt.title('Porcentaje de comentarios a favor, en contra o neutrales al proyecto')
plt.show()
```

```python
df_sin_duplicados = df.drop_duplicates(subset=['autor'])
counts = df_sin_duplicados['Bando'].value_counts()

# Mapear las etiquetas de las barras a los valores deseados
etiquetas = {
    'A': 'A favor',
    'C': 'En contra',
    'N': 'Neutrales'
}

# Reemplazar las etiquetas en el gráfico
counts.index = counts.index.map(etiquetas)


# Graficar la diferencia entre los valores 'A' y 'C'
plt.bar(counts.index, counts.values, color=['red', 'green', 'grey'])
plt.xlabel('Bando')
plt.ylabel('Número de ocurrencias')
plt.title('Usuarios a favor, en contra o neutrales al proyecto')

# Agregar el total de cada barra
for i, value in enumerate(counts.values):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.show()
```

```python
# Obtener los porcentajes de cada bando
percentages = df_sin_duplicados['Bando'].value_counts(normalize=True) * 100

etiquetas = {
    'C': 'En contra',
    'A': 'A favor',
    'N': 'Neutrales'
}
# Definir los colores para cada bando
colors = {'A': 'green', 'C': 'red', 'N': 'grey'}

# Crear el gráfico de torta con colores personalizados
plt.pie(percentages, labels=etiquetas.values(), autopct='%1.1f%%', colors=[colors[key] for key in percentages.index])
plt.title('Porcentaje de usuarios a favor o en contra del proyecto')
plt.show()
```
