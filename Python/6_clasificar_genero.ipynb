```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('comentarios.csv')
df = df[df['Género'] != 'X']
df = df[df['Género'] != 'N']
df = df.drop_duplicates(subset=['autor'])
```

```python
counts = df['Género'].value_counts()

# Definir los colores para cada valor del bando
colors = ['blue' if genero == 'M' else 'red' if genero == 'F' else 'grey' for genero in counts.index]

# Graficar la diferencia entre los valores 'A' y 'C'
plt.bar(counts.index, counts.values, color=colors)
plt.xlabel('Bando')
plt.ylabel('Número de ocurrencias')
plt.title('Usuarixs masculinos y femeninos')

# Agregar el total de cada barra
for i, value in enumerate(counts.values):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.show()
```

```python
# Obtener los porcentajes de cada bando
percentages = df['Género'].value_counts(normalize=True) * 100

# Definir los colores para cada bando
colors = {'M': 'blue', 'F': 'red'}

# Crear el gráfico de torta con colores personalizados
plt.pie(percentages, labels=percentages.index, autopct='%1.1f%%', colors=[colors[key] for key in percentages.index])
plt.title('Porcentaje de usuarixs masculinos y femeninos')
plt.show()
```
